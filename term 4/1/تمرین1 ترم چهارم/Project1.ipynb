## Import Modules

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

## Read File CSV 

```python
df = pd.read_csv('data.csv')

df.head()
```

```python
df.describe()
```

## Calculate the Mean and Median

```python
mean_value = df[['Age','Weight']].mean()
print(mean_value)
#=================
median_value = df[['Age','Weight']].median()
print(median_value)
```

## Set the Mode of Height,Gender,Education

```python
mode_value = df[['Height','Gender','Education']].mode()
print(mode_value)
#=====================
```

## Variance and STD Deviation

```python
variance_value = df[['Age','Height','Weight']].var()
print(variance_value)
#=============================
std_deviation_value = df[['Age','Height','Weight']].std()
print(std_deviation_value)
```

## Data Range for Age,Height,Weight

```python
data_range = np.ptp(df[['Age','Height','Weight']])
print(data_range)
```

## Quartiles for Age,Height,Weight

```python
age_quartiles = np.percentile(df['Age'],[25,50,75])
Height_quartiles = np.percentile(df['Height'],[25,50,75])
Weight_quartiles = np.percentile(df['Weight'],[25,50,75])
print(f"Age_Quartile={age_quartiles}")
print(f"Height_Quartile={Height_quartiles}")                                    
print(f"Weight_Quartile={age_quartiles}")                                    
```

## IQR for Age,Height,Weight

```python
age_iqr = np.percentile(df['Age'],75) - np.percentile(df['Age'],25)
Height_iqr = np.percentile(df['Height'],75) - np.percentile(df['Height'],25)
Weight_iqr = np.percentile(df['Weight'],75) - np.percentile(df['Weight'],25)
print(f"Age_IQR={age_iqr}")
print(f"Height_IQR={Height_iqr}")
print(f"Weight_IQR={Weight_iqr}")
```

## Bar Chart for Income_Monthly

```python
df['mean_income'] = df[['March income',
           'April income',
           'May income',
           'June income',
           'July income',
           'August income',
           'September income',
           'October income',
           'November income',
           'December income']].mean(axis=1)
df

plt.figure(figsize = (15,10))
plt.bar(df['Name'],df['mean_income'],color = 'midnightblue')
plt.xlabel('Name',fontsize = 20)
plt.ylabel('income_monthly',fontsize = 20)
plt.title('Average Income Chart',fontsize = 20)
plt.show()
```

## Concat the Name & Monthly_income

```python
income = df[['Name',
             'January income', 
             'February income',
             'March income',
             'April income',
             'May income',
             'June income',
             'July income',
             'August income',
             'September income',
             'October income',
             'November income',
             'December income']]
income
```

```python
plt.figure(figsize=(16,8))
plt.title('Income change over 12 month',fontsize = 20)
plt.xlabel('Month',fontsize = 15)
plt.ylabel('Income',fontsize = 15)

for index,row in income.iterrows():
    name = row['Name']
    income_values = row[1:]
    plt.plot(income_values,label = name , marker = 'o' , linestyle = '--')
    
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.xticks(range(12),months)
plt.legend(fontsize = 15)
plt.tight_layout()
plt.show()
```
