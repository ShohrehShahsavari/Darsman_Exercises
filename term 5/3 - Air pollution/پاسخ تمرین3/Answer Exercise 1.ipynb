```python
def clean_and_convert(df):
    # تبدیل ستون Date به فرمت datetime
    df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y', errors='coerce')
    # استخراج ساعت از ستون Time
    df['Hour'] = df['Time'].str.split('.').str[0].astype(float)
    # پیدا کردن ستون‌های عددی که به صورت object درآمده‌اند
    for col in df.select_dtypes(include=['object']).columns:
        # بررسی اینکه آیا ستون مربوطه حاوی کاما است
        if df[col].str.contains(',', na=False).any():
            # جایگزین کردن کاما با نقطه
            df[col] = df[col].str.replace(',', '.', regex=False)
            # تبدیل به عدد
            df[col] = pd.to_numeric(df[col], errors='coerce')

    return df

# فرض می‌کنیم df_copy داده‌تان باشد
df_copy = clean_and_convert(df_copy)
```

```python
def impute_missing_values(df, cols, method="mean"):
    df_filled = df.copy()  # جلوگیری از تغییر مستقیم روی df اصلی
    for col in cols:
        if df_filled[col].isnull().sum() > 0:
            if method == "mean":
                df_filled[col] = df_filled[col].fillna(df_filled[col].mean())
            elif method == "median":
                df_filled[col] = df_filled[col].fillna(df_filled[col].median())
            elif method == "ffill":
                df_filled[col] = df_filled[col].fillna(method="ffill")
            elif method == "bfill":
                df_filled[col] = df_filled[col].fillna(method="bfill")
    return df_filled
```

```python
def plot_scatter_separately(df, target_variable):
    # اطمینان از وجود متغیر هدف در دیتافریم
    if target_variable not in df.columns:
        raise ValueError(f"متغیر هدف '{target_variable}' در دیتافریم یافت نشد.")

    # انتخاب ویژگی‌های عددی
    numeric_cols = df.select_dtypes(include=["number"]).columns
    numeric_cols = numeric_cols[numeric_cols != target_variable]  # حذف متغیر هدف از لیست ویژگی‌ها

    # رسم نمودارهای پراکندگی به‌صورت جداگانه برای هر ویژگی
    for col in numeric_cols:
        plt.figure(figsize=(7, 5))
        sns.scatterplot(x=df[col], y=df[target_variable])
        plt.xlabel(col)
        plt.ylabel(target_variable)
        plt.title(f"Scatter Plot: {col} vs {target_variable}")
        plt.show()

plot_scatter_separately(df_copy, 'CO(GT)')
```

```python
from scipy.stats.mstats import winsorize
```

```python
def winsorize_and_handle_missing(df, cols, limits=[0.05, 0.05]):
    # کپی کردن دیتافریم برای جلوگیری از تغییر داده‌های اصلی
    df_filled = df.copy()

    for col in cols:
        if col in df_filled.columns:
            df_filled[col] = winsorize(df_filled[col], limits=limits)

    return df_filled
```

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectKBest, mutual_info_regression
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
```

```python
def extract_date_features(df):
    """
    ویژگی‌های زمانی (سال، ماه، روز) را از ستون 'Date' استخراج و آن را حذف می‌کند.

    این تابع با تبدیل ستون 'Date' به ویژگی‌های عددی جداگانه (سال، ماه، روز) دیتاست را
    غنی می‌کند تا در مدل‌ها قابل استفاده باشد. ابتدا بررسی می‌کند که ستون 'Date' وجود
    دارد یا نه؛ اگر وجود نداشته باشد، پیامی چاپ می‌کند و دیتافریم اصلی را بدون تغییر
    بازمی‌گرداند. پس از استخراج، ستون 'Date' برای جلوگیری از تکرار حذف می‌شود.
    """
    df = df.copy()
    if "Date" not in df.columns:
        print("ستون 'Date' در دیتافریم وجود ندارد. استخراج ویژگی‌های زمانی انجام نشد.")
        return df
    df["Year"] = df["Date"].dt.year
    df["Month"] = df["Date"].dt.month
    df["Day"] = df["Date"].dt.day
    return df.drop(columns=["Date"])


def remove_multicollinearity(X, thresh=20):
    """
    ویژگی‌هایی با همبستگی چندجمله‌ای بالا را بر اساس فاکتور تورم واریانس (VIF) حذف می‌کند.

    این تابع با استفاده از معیار VIF، که میزان تورم واریانس ضریب رگرسیون به دلیل
    همبستگی را اندازه‌گیری می‌کند، ویژگی‌های با همبستگی بالا را شناسایی و حذف می‌کند.
    ویژگی‌هایی که VIF آن‌ها بیشتر از آستانه (پیش‌فرض 10) باشد، به‌صورت تکراری حذف
    می‌شوند تا همه ویژگی‌های باقی‌مانده VIF قابل قبولی داشته باشند. مقادیر گمشده با
    میانگین پر می‌شوند تا محاسبات VIF انجام شود.
    """
    X = X.copy()
    X = X.fillna(X.mean())
    X = add_constant(X)
    while True:
        vif_data = pd.DataFrame()
        vif_data["Feature"] = X.columns
        vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
        max_vif = vif_data["VIF"].max()
        if max_vif > thresh:
            feature_to_drop = vif_data.loc[vif_data["VIF"].idxmax(), "Feature"]
            print(f"حذف ویژگی به دلیل Multicollinearity (VIF={max_vif:.2f}): {feature_to_drop}")
            X = X.drop(columns=[feature_to_drop])
        else:
            break
    print(f"ویژگی‌های باقی‌مانده پس از حذف Multicollinearity: {list(X.columns)}")
    return X


def mutual_info_selector(X, y, k=10):
    """
    k ویژگی برتر را بر اساس اطلاعات متقابل با متغیر هدف انتخاب می‌کند.

    این تابع با استفاده از معیار اطلاعات متقابل (Mutual Information) وابستگی بین هر
    ویژگی و متغیر هدف را ارزیابی کرده و k ویژگی برتر که بیشترین اطلاعات را ارائه
    می‌دهند را انتخاب می‌کند. داده‌ها به‌گونه‌ای تبدیل می‌شوند که فقط ویژگی‌های
    انتخاب‌شده باقی بمانند و امتیاز هر ویژگی برای درک اهمیت آن چاپ می‌شود.

    پارامترها:
        X (pandas.DataFrame): دیتافریم ورودی شامل ویژگی‌ها.
        y (pandas.Series): متغیر هدف.
        k (int, اختیاری): تعداد ویژگی‌های برتر برای انتخاب. پیش‌فرض 10 است.
    """
    selector = SelectKBest(score_func=mutual_info_regression, k=k)
    X_new = selector.fit_transform(X, y)
    selected_features = X.columns[selector.get_support()]
    scores = selector.scores_[selector.get_support()]
    print(f"ویژگی‌های انتخاب شده با Mutual Information: {list(selected_features)}")
    print(f"امتیاز Mutual Information هر ویژگی: {dict(zip(selected_features, scores))}")
    return pd.DataFrame(X_new, columns=selected_features)


def feature_engineering_and_selection(df, target_column, k=5, vif_threshold=10):
    """
    فرایند کامل مهندسی ویژگی و انتخاب ویژگی را اجرا می‌کند.

    این تابع کل فرآیند آماده‌سازی دیتاست برای مدل‌سازی را مدیریت می‌کند، از جمله
    استخراج ویژگی‌های زمانی، حذف ستون‌های غیرعددی، حذف ویژگی‌های با همبستگی
    چندجمله‌ای با استفاده از VIF، و انتخاب ویژگی‌های اطلاع‌رسان با Mutual Information.
    متغیر هدف حفظ شده و در نهایت به دیتافریم نهایی اضافه می‌شود تا برای آموزش
    مدل قابل استفاده باشد.

    پارامترها:
        df (pandas.DataFrame): دیتافریم ورودی شامل ویژگی‌ها و متغیر هدف.
        target_column (str): نام ستون متغیر هدف که باید پیش‌بینی شود.
        k (int, اختیاری): تعداد ویژگی‌های برتر برای انتخاب با Mutual Information. پیش‌فرض 10 است.
        vif_threshold (float, اختیاری): آستانه VIF برای حذف همبستگی چندجمله‌ای. پیش‌فرض 10 است.
    """
    y = df[target_column].reset_index(drop=True)
    df_features = df.drop(columns=[target_column])
    df_features = extract_date_features(df_features)
    df_features = remove_multicollinearity(df_features, thresh=vif_threshold)
    df_selected = mutual_info_selector(df_features, y, k)
    df_final = pd.concat([df_selected, y], axis=1)
    return df_final

# اجرای کد روی دیتافریم اصلی
df_processed = feature_engineering_and_selection(df_copy, target_column="CO(GT)", k=5, vif_threshold=50)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR

from sklearn.model_selection import (train_test_split,
                                     RandomizedSearchCV,
                                     learning_curve)
from sklearn.metrics import (mean_squared_error,
                             r2_score,
                             mean_absolute_error)

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
def prepare_data(df, target_column, test_size=0.2, random_state=42):
    """
    داده‌ها را به ویژگی‌های مستقل و وابسته تقسیم کرده و به داده‌های آموزشی و تست می‌شکند.

    این تابع ستون هدف را از دیتافریم جدا می‌کند، ویژگی‌های مستقل (X) و وابسته (y) را ایجاد می‌کند،
    سپس با استفاده از train_test_split داده‌ها را به بخش‌های آموزشی و تست (20% به صورت پیش‌فرض) تقسیم می‌کند.
    """
    # جدا کردن ویژگی‌ها و هدف
    X = df.drop(target_column, axis=1)
    y = df[target_column]
    # تقسیم داده‌ها به آموزش و تست
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test


def get_models():
    """
    مجموعه‌ای از مدل‌های پایه را برای آموزش بازمی‌گرداند.

    این تابع چهار مدل رگرسیون (Ridge، Lasso، SVR، GradientBoostingRegressor) را با
    تنظیم ثابت تصادفی ایجاد و در یک دیکشنری بازمی‌گرداند.
    """
    # تعریف مدل‌ها با تنظیم ثابت تصادفی
    models = {
        "Ridge": Ridge(random_state=42),
        "Lasso": Lasso(random_state=42),
        "SVR": SVR(),
        "GBM": GradientBoostingRegressor(random_state=42),
    }
    return models


def train_and_evaluate(models, X_train, X_test, y_train, y_test):
    """
    مدل‌ها را آموزش داده و با معیارهای مختلف ارزیابی می‌کند.

    این تابع هر مدل را با استفاده از یک پایپ لاین که شامل استانداردسازی و رگرسیون است،
    آموزش می‌دهد، پیش‌بینی انجام می‌دهد و معیارهای MAE، MSE و R² را برای داده‌های
    آموزشی و تست محاسبه و نمایش می‌دهد.
    """
    results = {}
    for name, model in models.items():
        # ایجاد pipeline با استانداردسازی و مدل
        pipeline = Pipeline([
            ("scaler", StandardScaler()),
            ("regressor", model)
        ])
        # آموزش مدل
        pipeline.fit(X_train, y_train)
        # پیش‌بینی روی داده‌های آموزشی و تست
        y_train_pred = pipeline.predict(X_train)
        y_test_pred = pipeline.predict(X_test)
        # محاسبه معیارهای ارزیابی
        mae_train = mean_absolute_error(y_train, y_train_pred)
        mse_train = mean_squared_error(y_train, y_train_pred)
        r2_train = r2_score(y_train, y_train_pred)
        mae_test = mean_absolute_error(y_test, y_test_pred)
        mse_test = mean_squared_error(y_test, y_test_pred)
        r2_test = r2_score(y_test, y_test_pred)
        # ذخیره نتایج
        results[name] = {
            "Train MAE": mae_train, "Test MAE": mae_test,
            "Train MSE": mse_train, "Test MSE": mse_test,
            "Train R²": r2_train, "Test R²": r2_test
        }
        # نمایش نتایج هر مدل
        print(f"{name}:")
        print(f"Train MAE = {mae_train:.4f}, Test MAE = {mae_test:.4f}")
        print(f"Train MSE = {mse_train:.4f}, Test MSE = {mse_test:.4f}")
        print(f"Train R² = {r2_train:.4f}, Test R² = {r2_test:.4f}")
        print("=" * 50)
    # نمایش نتایج نهایی
    print("\nنتایج نهایی:", results)
    return results


def plot_learning_curve(model, X, y, model_name):
    """
    نمودار Learning Curve را برای مدل مشخص‌شده رسم می‌کند.

    این تابع با استفاده از learning_curve، خطا را برای اندازه‌های مختلف داده‌های آموزشی
    محاسبه کرده و میانگین و انحراف معیار آن را برای داده‌های آموزشی و تست نمایش می‌دهد.
    """
    # محاسبه نقاط Learning Curve
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, y, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10)
    )

    # محاسبه میانگین و انحراف معیار
    train_mean = -train_scores.mean(axis=1)
    test_mean = -test_scores.mean(axis=1)
    train_std = train_scores.std(axis=1)
    test_std = test_scores.std(axis=1)

    # نمودار
    plt.figure(figsize=(7, 5))
    plt.plot(train_sizes, train_mean, label=f'{model_name} Train', color='blue')
    plt.plot(train_sizes, test_mean, label=f'{model_name} Test', color='red')
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='blue')
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color='red')
    plt.title(f'Learning Curve for {model_name}')
    plt.xlabel('Training Size')
    plt.ylabel('Mean Squared Error')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# تقسیم داده‌ها و آماده‌سازی مدل‌ها
X_train, X_test, y_train, y_test = prepare_data(df_processed, 'CO(GT)')
models = get_models() #دریافت مدل‌ها

#آموزش و ارزیابی مدل‌ها
results = train_and_evaluate(models, X_train, X_test, y_train, y_test)

#رسم نمودار Learning Curve برای هر مدل
for name, model in models.items():
    pipeline = Pipeline([
        ("scaler", StandardScaler()),
        ("regressor", model)
    ])

    plot_learning_curve(pipeline, X_train, y_train, name)
    print("=" * 70)
```
